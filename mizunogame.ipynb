```python
import pygame
import random
from pygame.locals import *
import sys
def main():
    pygame.init()                                 # Pygameの初期化
    screen = pygame.display.set_mode((450,450))  # 800*600の画面
    px=50
    py=50
    px2=50
    py2=350
    px3=200
    py3=350
    px4=300
    py4=350
    hx=350
    hy=200
    ex=300
    ey=300
    SIZE=50
    g1=0
    r1=0
    r=random.randint(1,4)
    r2=random.randint(1,4)
    r3=random.randint(1,4)
    r4=random.randint(1,4)
    ar=[["c","c","c","c","c","c","c","c","c"],
        ["c","s","c","c","c","c","c","c","c"],
        ["c","c","c","c","c","c","c","c","c"],
        ["c","c","c","c","c","c","c","c","c"],
        ["c","c","c","c","c","c","c","c","c"],
        ["c","c","c","c","c","c","c","c","c"],
        ["c","c","c","c","c","c","c","c","c"],
        ["c","c","c","c","c","c","c","c","c"],
        ["c","c","c","c","c","c","c","c","c"]]
    gPng = pygame.image.load("hpatimon.png").convert_alpha()   #画像を読み込む
    ePng = pygame.image.load("patimon.png").convert_alpha()   #画像を読み込む
    pPng = pygame.image.load("pac.png").convert_alpha()   #画像を読み込む
    p2Png = pygame.image.load("pac2.png").convert_alpha()   #画像を読み込む
    p3Png = pygame.image.load("pac3.png").convert_alpha()   #画像を読み込む
    p4Png = pygame.image.load("pac4.png").convert_alpha()   #画像を読み込む
    while True:
        screen.fill((255,255,255))
        for g in range(9):
            for r in range(9):
                if ar[g][r]=="n":
                    pygame.draw.rect(screen, (255,0,0), Rect(SIZE*r,SIZE*g,SIZE,SIZE))
        screen.blit(gPng ,Rect(hx,hy,50,50))#オブジェクトを指定の座標に表示
        screen.blit(ePng ,Rect(ex,ey,50,50))#オブジェクトを指定の座標に表示
        screen.blit(pPng ,Rect(px,py,50,50))#オブジェクトを指定の座標に表示
        screen.blit(p2Png ,Rect(px2,py2,50,50))#オブジェクトを指定の座標に表示
        screen.blit(p3Png ,Rect(px3,py3,50,50))#オブジェクトを指定の座標に表示
        screen.blit(p4Png ,Rect(px4,py4,50,50))#オブジェクトを指定の座標に表示

#             if ar[2][3]=="n":
#                 pygame.draw.rect(screen, (255,0,0), Rect(361,241,120,120))
#             if ar[3][3]=="n":
#                 pygame.draw.rect(screen, (255,0,0), Rect(361,361,120,120))
#             if ar[2][2]=="n":
#                 pygame.draw.rect(screen, (255,0,0), Rect(241,241,120,120))
#             if ar[4][1]=="n":
#                 pygame.draw.rect(screen, (255,0,0), Rect(121,481,120,120))
        pygame.display.update()
        g1=px/50
        r1=py/50
        if r==1 and ar[g1][r1+1]=="c":
            px=px+50
        if r==2 and ar[g1][r1-1]=="c":
            px=px-50
        if r==3 and ar[g1-1][r1]=="c":
            py=py+50
        if r==4 and ar[g1+1][r1]=="c":
            py=py-50
        if r2==1 :
            px2=px2+50
        if r2==2:
            px2=px2-50
        if r2==3:
            py2=py2+50
        if r2==4:
            py2=py2-50
        if r3==1:
            px=px+50
        if r3==2:
            px=px-50
        if r3==3:
            py=py+50
        if r3==4:
            py=py-50
        if r4==1:
            px=px+50
        if r4==2:
            px=px-50
        if r4==3:
            py=py-50
    

        # イベント処理
        for event in pygame.event.get():  # イベントを取得
            if event.type == QUIT:        # 閉じるボタンが押されたら
                pygame.quit()             
                sys.exit()                # 終了
            if event.type == KEYDOWN:
                if event.key==K_LEFT:
                  hx -=50  # 横方向の速度
                if event.key==K_RIGHT:
                  hx +=50  # 横方向の速度
                if event.key==K_UP:
                  hy -=50 # 横方向の速度
                if event.key==K_DOWN:
                  hy +=50  # 横方向の速度
if __name__ == "__main__":
    main()
```
